# Merging LAP, P5V, GDP

## Functions

```python
def read_file(file_string):
    f = open(file_string, "r")
    df = []
    headers = f.readline()
    line = f.readline()
    while line:
        line = line.strip().split(",")
        df.append(line)
        line = f.readline()
    f.close()
    return df
```

```python
def get_instances_in(df, country, year, col, element):
    n = 0
    for i in range(len(df)):
        if(str(country) == str(df[i][0])):
            if(str(year) == str(df[i][2])):
                n = n + (df[i][col].lower()).count(element)
    return n
```

```python
def get_unique_strings(df, col):
    line = ""
    for i in range(len(df)):
        line = line + ";" + str(df[i][col])
    line = line.split(";")
    for i in range(len(line)):
        line[i] = (line[i]).strip()
        line[i] = (line[i]).lower()
    line = [string for string in line if string != ""]
    import numpy as np
    linenp = np.array(line)
    unique = np.unique(linenp)
    return unique
```

## Read in files

### Laws and Policies

```python
lap = read_file("LAP.csv")
print(lap[0])
```

```python
new = []

#geography , geography iso, y, type, frameworks, responses, instruments, document types, natural hazards, keywords, sectors
rel = [2, 3, 13, 1, 4, 5, 6, 7, 8, 9, 10]

for i in range(len(lap)):
    row = []
    for j in range(len(rel)):
        row.append(lap[i][rel[j]])
    new.append(row)
    
lap = new
```

### GDP

```python
gdp = read_file("GDP.csv")
print(gdp[0])
```

```python
new = []
for i in range(len(gdp)):
    for j in range(len(gdp[i]) - 4):
        row = []
        row.append(gdp[i][0])            #country
        row.append(str(gdp[i][1]))       #country code
        row.append(str(1960+j))          #year
        row.append(str(gdp[i][j+4]))     #gdp
        if( int(row[2]) > 1979 and int(row[2]) < 2021): #only take 1980 - 2020
            new.append(row)
            
gdp = new
```

```python
import pandas as pd
pddf_gdp = pd.DataFrame(gdp, columns=['country', 'scode', 'year', 'GDPpc'])
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pddf_gdp.head(10)
```

## Creating new LAP table

```python
#new columns from splitting lap columns

ls_uniques = []
for i in range(len(lap[0])-3):
    ls_uniques.append(get_unique_strings(lap,i+3))
```

```python
#get_instances_in(df, country, year, col, element)
print(get_instances_in(lap, "Afghanistan", 2014, 3, "executive"))
```

```python
ls_none = []
for i in range(len(p5v[0])-3):
    ls_none.append('')
print(ls_none)
```

```python
new = []

for i in range(len(gdp)):
    row = []
    row = gdp[i]
    
    match = False
    for j in range(len(p5v)):
        if(p5v[j][3] == gdp[i][1] and p5v[j][5] == gdp[i][2]):
            match = True
            row = row + p5v[j][:3] + p5v[j][6:] 
    if(match == False):
        row = row + ls_none
        
    for j in range(len(ls_uniques)):
        for k in range(len(ls_uniques[j])):
            row.append(get_instances_in(lap, gdp[i][0], gdp[i][2], j+3, ls_uniques[j][k]))
    new.append(row)
    
    if(i % 100 == 0):
        print(str(round(i/(len(gdp))*100,3))+"%")
```

```python
lap_headers = []
for i in range(len(ls_uniques)):
    for j in range(len(ls_uniques[i])):
        lap_headers.append(ls_uniques[i][j])
        
p5v_headers=['p5', 'cyear', 'ccode', 'flag', 'fragment', 'democ', 'autoc', 'polity',
             'polity2', 'durable', 'xrreg', 'xrcomp', 'xropen', 'xconst', 'parreg', 'parcomp', 'exrec',
             'exconst', 'polcomp', 'prior', 'emonth', 'eday', 'eyear', 'eprec', 'interim', 'bmonth', 'bday',
             'byear', 'bprec', 'post', 'change', 'd5', 'sf', 'regtrans']
```

```python
import pandas as pd
pddf_new = pd.DataFrame(new, columns=[['country', 'scode', 'year', 'GDPpc'] + p5v_headers + lap_headers]) 

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pddf_new.head(120)
```
